```python
# ABBY
# determine background - based on olivia's version (2 images for pairwise comparison)

from matplotlib import pyplot as plt
import cv2
import numpy as np

raw_images = []

raw_images.append('/content/13_07_42_725696.jpg')
raw_images.append('/content/13_07_45_211842.jpg')

# helper function to read in an image, and grayscale
def read_in(image):
  img = cv2.imread(image)
  return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# helper function that prints an image
def print_image(image):
  print(image[1400][230])
  plt.figure(figsize=(50,30))
  plt.imshow(background)
  plt.title('background')
  plt.show()

updated_images = []
for image in raw_images:
  updated_image = read_in(image)
  updated_images.append(updated_image)

height, width = updated_images[0].shape
background = np.ones(updated_images[0].shape[:2], dtype="uint8")

threshold = 8

for row in range(height):
  for col in range(width):
    to_compare = []
    for i in range(len(updated_images)):
      to_compare.append(updated_images[i][row][col])

    difference = np.absolute(int(to_compare[0]) - int(to_compare[1]))
    temp_array = np.array([to_compare[0], to_compare[1]])

    """ counter for the number of pixels that are categorized as dark spots in the
    background image (dust spots or stuck organisms """
    num_pixels_spots = 0

    if (difference > threshold):
      # one has a dark spot, the other does not (this would indicate a flowing organism)
      background[row][col] = max(temp_array)
    else:
      # both images have similar value pixels, so we set the background image to the average of the two
      background[row][col] = np.average(temp_array)
      num_pixels_spots += 1

print_image(background)
percent_spots = (num_pixels_spots/(width*height))*100

print(percent_spots)

if (percent_spots >= 3):
  print("time to clean scope")
```
